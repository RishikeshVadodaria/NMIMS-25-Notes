```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_box(ax, x, y, width, height, text, color='lightblue', text_color='black'):
    ax.add_patch(patches.Rectangle((x, y), width, height, edgecolor='black', facecolor=color, lw=1.5))
    ax.text(x + width/2, y + height/2, text, fontsize=10, ha='center', va='center', color=text_color, fontweight='bold')

fig, ax = plt.subplots(figsize=(8, 10))
ax.set_xlim(0, 10)
ax.set_ylim(0, 12)
ax.axis('off')

# Define block positions and dimensions
block_width = 3
block_height = 0.8

# Define layers
layers = [
    (3.5, 11, 'WAP', 'orange'),
    (3.5, 10.2, 'TCP/UDP', 'orange'),
    (3.5, 9.4, 'IP', 'orange'),
    (3.5, 8.6, 'PPP', 'orange'),
    (3.5, 7.8, 'AT Command', 'lightcoral'),
    (3.5, 7.0, 'SDP', 'lightcoral'),
    (3.5, 6.2, 'RFCOMM', 'lightgreen'),
    (3.5, 5.4, 'L2CAP', 'lightblue'),
    (3.5, 4.6, 'HCI', 'gold'),
    (3.5, 3.8, 'LMP', 'lightblue'),
    (3.5, 3.0, 'Baseband', 'blue'),
    (3.5, 2.2, 'Bluetooth Radio', 'blue'),
    (7.5, 7.8, 'Audio', 'purple')
]

# Draw the blocks
for x, y, label, color in layers:
    draw_box(ax, x, y, block_width, block_height, label, color)

# Draw connecting lines
connections = [
    ((5, 11), (5, 10.8)), ((5, 10.2), (5, 10)), ((5, 9.4), (5, 9.2)), ((5, 8.6), (5, 8.4)),
    ((5, 8.6), (5, 8.2)), ((5, 7.8), (5, 7.6)), ((5, 7.0), (5, 6.8)), ((5, 6.2), (5, 6.0)),
    ((5, 5.4), (5, 5.2)), ((5, 4.6), (5, 4.4)), ((5, 3.8), (5, 3.6)), ((5, 3.0), (5, 2.8)),
    ((5, 2.2), (5, 2.0)), ((5, 5.4), (8, 8)), ((8, 8), (8, 7.8))
]

for (x1, y1), (x2, y2) in connections:
    ax.plot([x1, x2], [y1, y2], 'k-', lw=1.5)

# Title
ax.text(5, 1.2, 'Bluetooth Protocol Stack', fontsize=12, ha='center', fontweight='bold', bbox=dict(facecolor='white', edgecolor='black'))

plt.show()
```
