## **Task 4:** Apply a set of filters to highlight edgesin different directions

```python
img2 = cv2.imread("elephant.jfif")
```

```python
img2_g = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
filters = []
num_filters = 16
SIZE = 31
SIGMA = 2
LAMBDA = 5
GAMMA = 0.5
PSI = 0

for THETA in np.arange(0, np.pi, np.pi / num_filters):
    f = cv2.getGaborKernel((SIZE, SIZE), SIGMA, THETA, LAMBDA, GAMMA, PSI)
    f = f / f.sum()
    filters.append(f)
```

```python
img2_f = np.zeros_like(img2_g)

for f in filters:
  img = cv2.filter2D(img2_g, -1, f)
  np.maximum(img2_f, img, img2_f)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.axis('off')
plt.imshow(img2_g, cmap='gray')
plt.title('Hidden Image')

plt.subplot(1, 2, 2)
plt.axis('off')
plt.imshow(img2_f, cmap='gray')
plt.title('Filtered Image')
```
